```python
import pandas as pd
import zipfile
from pathlib import Path
```

```python

# -------- SETTINGS --------
data_dir = Path("C:\\Users\\zollner\\OneDrive - epfl.ch\\Documents\\SGR TZ\\Data\\Raw")  # directory containing zip files
zip_prefix = "movement-distribution"
country_code = "TZA"
country_cols = ["country"]  # possible column names
output_file = "TZA_merged.csv"
# --------------------------

dfs = []

def find_country_column(df, candidates):
    """Return the first matching country column name."""
    for col in candidates:
        if col in df.columns:
            return col
    return None


for zip_path in data_dir.glob(f"{zip_prefix}*.zip"):
    print(f"Processing {zip_path.name}")

    with zipfile.ZipFile(zip_path, "r") as z:
        for file_name in z.namelist():

            # Skip non-CSV files
            if not file_name.lower().endswith(".csv"):
                continue

            try:
                with z.open(file_name) as f:
                    df = pd.read_csv(f)

                country_col = find_country_column(df, country_cols)
                if country_col is None:
                    print(f"  No country column in {file_name}")
                    continue

                df_tza = df[df[country_col] == country_code]

                if not df_tza.empty:
                    df_tza["source_zip"] = zip_path.name
                    df_tza["source_file"] = Path(file_name).name
                    dfs.append(df_tza)

            except Exception as e:
                print(f"  Error reading {file_name}: {e}")


# Merge everything
if dfs:
    final_df = pd.concat(dfs, ignore_index=True)
    final_df.to_csv(output_file, index=False)
    print(f"\nSaved {len(final_df)} rows to {output_file}")
else:
    print("No TZA data found.")
```

```python
# Load data
data = pd.read_csv("C:\\Users\\zollner\\OneDrive - epfl.ch\\Documents\\SGR TZ\\Data\\Raw\\TZA_merged.csv")
```

```python
# Delete duplicates
data = data.drop_duplicates(subset=['gadm_id', 'ds', 'home_to_ping_distance_category', 'distance_category_ping_fraction'])

# Do share sum up to one

check = (
    data.groupby(['gadm_id','ds'])['distance_category_ping_fraction']
      .sum()
      .reset_index()
)

check['distance_category_ping_fraction'].describe()
```

```python
daily = (
    data.groupby(['ds','home_to_ping_distance_category'])['distance_category_ping_fraction']
      .mean()
      .reset_index()
)

daily['ds'] = pd.to_datetime(daily['ds'], format="%Y-%m-%d")

event_date = pd.to_datetime("2024-08-01", format="%Y-%m-%d")
daily['event_time'] = (daily['ds'] - event_date).dt.days
```

```python
import matplotlib.pyplot as plt

plt.figure()

for b in daily['home_to_ping_distance_category'].unique():
    tmp = daily[daily['home_to_ping_distance_category'] == b]
    plt.plot(tmp['event_time'], tmp['distance_category_ping_fraction'], label=b)

plt.axvline(0)
plt.xlabel("Days relative to SGR opening")
plt.ylabel("Share of pings")
plt.legend()
plt.title("Distance from home  Tanzania")
plt.show()
```

```python
wide_daily = daily.pivot(
    index='event_time',
    columns='home_to_ping_distance_category',
    values='distance_category_ping_fraction'
).sort_index()

plt.figure()
plt.stackplot(
    wide_daily.index,
    wide_daily.T,
    labels=wide_daily.columns
)
plt.axvline(0)
plt.legend(loc='upper left')
plt.xlabel("Days relative to SGR opening")
plt.ylabel("Share")
plt.show()
```

```python
sgr_corridor = [
    'Ilala', 'Temeke', 'Kinondoni',   # Dar es Salaam
    'Morogoro Urban', 'Morogoro Rural',
    'Dodoma Urban'
]
data['sgr_corridor'] = data['gadm_name'].isin(sgr_corridor)
data['ds'] = pd.to_datetime(data['ds'], errors='coerce')


corridor_daily = (
    data[data['home_to_ping_distance_category'] == "[10, 100)"]
    .groupby(['ds','sgr_corridor'])['distance_category_ping_fraction']
    .mean()
    .reset_index()
)

for val, label in [(True,"SGR corridor"), (False,"Other areas")]:
    tmp = corridor_daily[corridor_daily['sgr_corridor'] == val]
    plt.plot(tmp['ds'], tmp['distance_category_ping_fraction'], label=label)

plt.axvline(pd.to_datetime("01/08/2024", format="%d/%m/%Y"))
plt.legend()
plt.title("Medium-distance mobility share")
plt.show()

#restrict x axis to one month before and after SGR opening
plt.figure()
for val, label in [(True,"SGR corridor"), (False,"Other areas")]:
    tmp = corridor_daily[corridor_daily['sgr_corridor'] == val]
    plt.plot(tmp['ds'], tmp['distance_category_ping_fraction'], label=label)    
plt.xlim(pd.to_datetime("01/04/2024", format="%d/%m/%Y"), pd.to_datetime("30/12/2024", format="%d/%m/%Y"))
plt.axvline(pd.to_datetime("01/08/2024", format="%d/%m/%Y"))
plt.legend()


```
